```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn
%matplotlib inline
```

```python
train_file = '../data/train.csv'
test_file = '../data/test.csv'
```

```python
df = pd.read_csv(train_file)
df
```

```python
# Age, Sex, Name, Survived, Pclass, SibSp, Parch, Fare, Cabin, Embarked
cols_to_keep = [ 'Survived', 'Pclass', 'Name', 'Sex', 'Age', 'SibSp',
       'Parch', 'Fare', 'Cabin', 'Embarked']
df[cols_to_keep]
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['Age'],how='any',inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Cabin'].unique()
```

```python
def parse_cabin(s):
    if type(s) != str:
        return str(s)
    else:
        deck = s[0]
        return deck
```

```python
df['Cabin'].apply(parse_cabin)
```

```python
def categorize_deck(deck):
    if deck == 'nan':
        return 0
    else:
        return ord(deck) - 64
```

```python
df['Deck'] = df['Cabin'].apply(parse_cabin).apply(categorize_deck)
```

```python
df['Deck'].value_counts()
```

```python
df
```

```python
number_in_decks=[12,45,51,31,30,11,4,1]
fraction_in_decks = [x/sum(number_in_decks) for x in number_in_decks]
fraction_in_decks
```

```python
replacement_decks=[np.random.choice([1,2,3,4,5,6,7,20], p=fraction_in_decks) for x in range(529)]
```

```python
df['Deck'][df['Deck']==0]=replacement_decks
```
